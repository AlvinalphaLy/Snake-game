```python
import matplotlib.pyplot as plt
from random import choice
%matplotlib inline   
# In Jupyter notebook if you want to show plot , you have to have this line : %matplotlib inline 

def trans_1 (p):
    x = p[0]
    y = p[1]
    x1 = 0.5 * x
    y1 = 0.5 * y
    return x1,y1
# p[0] vs p[1] trong bài này là tọa độ, coordinate at 1 for y, and 0 for x
def trans_2(p):
    x = p[0]
    y = p[1]
    x1 = 0.5 * x + 0.5
    y1 = 0.5 * y + 0.5
    return x1,y1

def trans_3 (p):
    x = p[0]
    y = p[1]
    x1 = 0.5 * x + 1
    y1 = 0.5 * y 
    return x1,y1

transformations = [trans_1 , trans_2 ,trans_3]
a1 = [0]
b1 = [0]
a,b = 0,0
# đưa trans_1 , trans_2 , trans_3 vào list để xuống dưới dùng choice chọn random trong 3 cái trans đó, choice import từ random.
for i in range (1000):
    trans = choice (transformations)
    a,b = trans((a,b))
    a1.append(a)
    b1.append(b)
# 
plt.rc('figure', figsize=(12,12))

plt.plot(a1,b1,'o') 
# CÂU LỆNH ĐỂ LƯU HÌNH ẢNH TRONG MÁY TÍNH BELOW, RUN RỒI VÀO Ổ C XEM: 
# plt.savefig ('my_figure.png')
# 'o' dấu chấm trên , với ko nên để comment line ngang với câu lệnh
```
