```python
import matplotlib.pyplot as plt
from random import choice
%matplotlib inline   
# In Jupyter notebook if you want to show plot , you have to have this line : %matplotlib inline 

def trans_1 (p):
    x = p[0]
    y = p[1]
    x1 = 0.5 * x
    y1 = 0.5 * y
    return x1,y1
# p[0] vs p[1] trong bi ny l ta , coordinate at 1 for y, and 0 for x
def trans_2(p):
    x = p[0]
    y = p[1]
    x1 = 0.5 * x + 0.5
    y1 = 0.5 * y + 0.5
    return x1,y1

def trans_3 (p):
    x = p[0]
    y = p[1]
    x1 = 0.5 * x + 1
    y1 = 0.5 * y 
    return x1,y1

transformations = [trans_1 , trans_2 ,trans_3]
a1 = [0]
b1 = [0]
a,b = 0,0
# a trans_1 , trans_2 , trans_3 vo list  xung di dng choice chn random trong 3 ci trans , choice import t random.
for i in range (1000):
    trans = choice (transformations)
    a,b = trans((a,b))
    a1.append(a)
    b1.append(b)
# 
plt.rc('figure', figsize=(12,12))

plt.plot(a1,b1,'o') 
# CU LNH  LU HNH NH TRONG MY TNH BELOW, RUN RI VO  C XEM: 
# plt.savefig ('my_figure.png')
# 'o' du chm trn , vi ko nn  comment line ngang vi cu lnh
```
